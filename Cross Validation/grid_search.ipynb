```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
X = df.drop('Survived', axis = 1)
y = df['Survived']

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(SVC(gamma = 'auto'), {
    'C' : [1, 10, 20],
    'kernel' : ['rbf', 'linear']
}, cv = 3, return_train_score=False)

clf.fit(X, y)
clf.best_params_
dir(clf)
```
