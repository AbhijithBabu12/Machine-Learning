```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

```python
np.random.seed(42)


study = np.random.uniform(0, 10, 100)

noise = np.random.normal(0, 5, 100)

score = 9.5 * study + 10 + noise


data = pd.DataFrame({
    'study': study,
    'score': score
})

print(data.head())

print("Correlation :", data['study'].corr(data['score']))


plt.scatter(data['study'], data['score'], color = 'blue')
m, b = np.polyfit(data['study'], data['score'], 1)
print(f"M :{m} B :{b}")
x_line = np.linspace(data['study'].min(), data['study'].max(), 100)
y_line = m * x_line + b

plt.plot(x_line, y_line, color='red',  label=f'y = {m:.2f}x + {b:.2f}')
plt.xlabel('Study Hours')
plt.ylabel('Score')
plt.title('Study Hours vs Score with Linear Regression Line')
plt.legend()
plt.grid(True)
```
