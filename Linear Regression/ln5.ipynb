```python
data = pd.read_csv('insurance.csv')
data
```

```python
plt.scatter(data['Age'], data['Premium'])
```

```python
print(data['Age'].corr(data['Premium']))
```

```python
x = data[['Age']]
y = data['Premium']

model = LinearRegression()
model.fit(x, y)
m = model.coef_[0]
b = model.intercept_
print("m : ", model.coef_[0])
print("b : ", model.intercept_)

from sklearn.metrics import r2_score

y_test = data['Premium']
x_val = data['Age']
y_pred = model.predict(x)
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
plt.scatter(data['Age'], data['Premium'])
x_line = np.linspace(data['Age'].min(), data['Age'].max(), 100)
plt.plot(x_line, m * x_line + b, color='red', label = f"y = {m:.2f}x + {b:.2f}")
plt.xlabel('Age')
plt.ylabel('Premium')
plt.legend()
plt.title(f'Linear Regression: Age vs Premium (RÂ² = {r2:.2f})')
```
